```python
df2.dropna(inplace=True)
df2.isnull().any()
```

```python
df2.dtypes[df2.dtypes == np.object]
cat_columns = list(df2.dtypes[df2.dtypes == np.object].index)
num_columns = list(df2.dtypes[df2.dtypes != np.object].index)
num_columns
```

```python
df2 = df2.fillna({'Flow ID':'', ' Source IP':'', ' Destination IP':'', ' Timestamp':'',' Label':''})
df2 = df2.fillna(0)
```

```python
# from sklearn.impute import SimpleImputer
# imp = SimpleImputer(missing_values=np.nan, strategy='mean')
# imp.fit(df2)
df2_cat = df2[cat_columns]
df2_num = df2.drop(cat_columns, axis = 1)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer()
df2_cat=vectorizer.fit_transform(df2_cat.values.ravel())
```

```python
len(vectorizer.get_feature_names())
```

```python
df2_cat = pd.DataFrame(df2_cat, columns = vectorizer.get_feature_names())
```

```python
df2_num = df2[num_columns]
df2_encoded = df2_cat
df2_encoded[num_columns] = df2_num.values
df2_encoded =df2_encoded[~df2_encoded.isin([np.nan, np.inf, -np.inf]).any(1)]
```

```python
import imblearn
from imblearn.over_sampling import SMOTE
oversample = SMOTE()
X, y = oversample.fit_resample(df2_encoded.iloc[:, :-1].values, df2_encoded.iloc[:, -1].values)
```
